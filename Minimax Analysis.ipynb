```python
import matplotlib.pyplot as plt
import numpy
```

```python
four_x = []
four_y = []
four_itr = []

with open('four.txt') as file:
    ctr = 0
    for line in file:        
        if line.strip() == 'n':
            ctr = 1
        else:
            four_itr += [ctr]
            ctr += 1
            num = line.strip().split(',')
            four_x += [int(num[0])]
            four_y += [float(num[1])]
```

```python
plt.plot(four_x, four_y, '.')
plt.show()
```

```python
plt.plot(four_itr, four_y, '.')
plt.show()
```

```python
four_weighted = [four_x[i] * four_itr[i] for i in range(len(four_x))]
plt.plot(four_weighted, four_y, '.')
plt.show()
```
